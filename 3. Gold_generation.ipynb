# IMPORTACIÓN DE LIBRERÍAS

```python
# Importación de las librerías a usar durante el proyecto.

import os

import numpy as np

import pandas as pd
```

# PARÁMETROS DE LAS RUTAS PARA EL ARCHIVO PLATA Y ORO

```python
# Rutas para el archivo Silver.

silver_folder_path = os.path.join(os.curdir,"Silver")

silver_file_name = "student_habits_performance_silver.parquet"

silver_file_path = os.path.join(silver_folder_path, silver_file_name)


# Rutas para el archivo Oro. 

gold_folder_path = os.path.join(os.curdir,"Gold")

gold_file_name = "student_habits_performance_gold.parquet"

gold_file_path = os.path.join(gold_folder_path, gold_file_name)
```

# LECTURA DEL ARCHIVO SILVER

```python
# Leemos el archivo parquet con los datos en silver previamente creado con la libreta Silver generation.

student_habits_performance_silver_df = pd.read_parquet(silver_file_path)

# Muestra las 5 primeras filas del dataframe para echar un primer vistazo.

student_habits_performance_silver_df.head() 
```

# TRANSFORMACIONES SILVER2GOLD

```python
student_habits_performance_silver2gold_df = student_habits_performance_silver_df
```

#### CATEGORIZACIÓN BINARIA

```python
# Con la siguiente función podemos categorizar notas de un valor numérico a formato "Suspende" y "Aprueba".

def categorizar_notas_binaria(score):
    if score < 50:
        return 'Suspenso'
    else:
        return 'Aprobado'

student_habits_performance_silver2gold_df['exam_score_categoric_binary'] = student_habits_performance_silver2gold_df['exam_score'].apply(categorizar_notas_binaria)
```

#### CATEGORIZACIÓN PROGRESIVA

```python
# Con la siguiente función podemos categorizar notas en valor numérico y pasarlas a formato "Suspenso", "Aprobado", etc.

def categorizar_notas_progresive(score):
    if score < 50:
        return 'Suspenso'
    elif score < 70:
        return 'Aprobado'
    elif score < 90:
        return 'Notable'
    else:
        return 'Sobresaliente'

student_habits_performance_silver2gold_df['exam_score_categoric'] = student_habits_performance_silver2gold_df['exam_score'].apply(categorizar_notas_progresive)
```

#### ELIMINACIÓN DE COLUMNAS INNECESARIAS

```python
# Eliminamos la columna timestamp ya que no la necesitamos.
student_habits_performance_silver2gold_df = student_habits_performance_silver2gold_df.drop(columns=['BronzeTimestamp'])
```

# GUARDADO DEL ARCHVIO ORO

```python
# Creamos el directorio del archivo oro en caso de que no existiese.

os.makedirs(gold_folder_path, exist_ok=True)

# Guardamos el dataset en un archivo .parquet

student_habits_performance_silver2gold_df.to_parquet(gold_file_path, index=False)
```
