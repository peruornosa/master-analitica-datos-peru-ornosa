# IMPORTS

```python
# Importación de las librerías a usar durante el proyecto.

import os

import pandas as pd

from datetime import datetime as dt
```

# PARÁMETROS DE LAS RUTAS PARA EL ARCHIVO BRONZE

```python
raw_data_path = os.path.join(os.curdir,"student_habits_performance (SUCIO).csv")

bronze_folder_path = os.path.join(os.curdir,"Bronze")

bronze_file_name = "student_habits_performance_bronze.parquet"

bronze_file_path = os.path.join(bronze_folder_path, bronze_file_name)
```

# PREPARACIÓN DEL ARCHIVO BRONCE

#### LECTURA DEL DATA SET DE ORIGEN EN CRUDO

```python
# Lo primero que vamos a hacer es leer el archivo en crudo para poder preparar posteriormente el dataset nivel bronce añadiendo un timestamp.

student_habits_rawdatset_df = pd.read_csv(raw_data_path)


# Representamos las primeras 5 filas para ver el aspecto del dataset.

student_habits_rawdatset_df.head()


# Visualizamos la información del dataset proporcionada por el metodo info().

student_habits_rawdatset_df.info()
```

### AÑADIR TIMESTAMP

```python
# Se añade el timestamp al dataframe del rawdataset para dejar listo lo que será el bronze dataset. Además lo visualizamos para confirmar que ha sido creectamente añadido.

student_habits_rawdatset_df["BronzeTimestamp"] = dt.now()

student_habits_rawdatset_df.head()
```

# GUARDAR ARCHIVO BRONZE DATASET

```python
# Creamos el directorio del archivo bronze en caso de que no existiese.

os.makedirs(bronze_folder_path, exist_ok=True)

# Guardamos el dataset en un archivo .parquet

student_habits_rawdatset_df.to_parquet(bronze_file_path, index=False)
```
